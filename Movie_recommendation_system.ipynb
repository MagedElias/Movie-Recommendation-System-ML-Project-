```python
# set up
import pandas as pd 
movie_ratings = pd.read_csv('C:/Users/COMPUMARTS/OneDrive/Desktop/Elevvo Pathways ML Internship/Tasks/movie_rec_sys.data',
    sep="\t",
    names=["user_id", "movie_id", "rating", "timestamp"])

movie_ratings
```

```python
movies = pd.read_csv('C:/Users/COMPUMARTS/OneDrive/Desktop/Elevvo Pathways ML Internship/Tasks/u.item',
    sep="|",
    encoding="latin-1",
    names=[
        "movie_id","title","release_date","video_release_date","IMDb_URL",
        "unknown","Action","Adventure","Animation","Children","Comedy","Crime",
        "Documentary","Drama","Fantasy","Film-Noir","Horror","Musical","Mystery",
        "Romance","Sci-Fi","Thriller","War","Western"
    ]
)
movies = movies[["movie_id", "title"]]
movies
```

```python
movie_ratings = movie_ratings.merge(movies, on="movie_id")
movie_ratings
```

```python
# user item matrix
user_item = movie_ratings.pivot_table(
    index="user_id",
    columns="movie_id",
    values="rating"
).fillna(0)
user_item
```

```python
# Similarity Model
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(user_item)
predicted_ratings = np.dot(user_similarity, user_item.values) / np.sum(user_similarity, axis=1).reshape(-1,1)
```

```python
# The actual recommendations 
import numpy as np
def recommend_movies(user_id, k=5):
    user_index = user_id - 1
    user_predictions = predicted_ratings[user_index]

    user_rated = user_item.iloc[user_index]
    user_predictions[user_rated > 0] = 0

    top_movie_ids = np.argsort(user_predictions)[-k:][::-1]
    return movies[movies.movie_id.isin(top_movie_ids)][["movie_id","title"]]
```

```python
# Evaluating but with using K value 
def precision_at_k(user_id, k=5):
    recs = recommend_movies(user_id, k)
    actual = movie_ratings[(movie_ratings.user_id == user_id) & (movie_ratings.rating >= 4)].movie_id.values

    hits = sum([1 for m in recs.movie_id if m in actual])
    return hits / k

item_similarity = cosine_similarity(user_item.T)
```
